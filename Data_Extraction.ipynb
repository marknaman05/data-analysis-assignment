```python
import requests
from bs4 import BeautifulSoup
URL = "https://insights.blackcoffer.com/ai-in-healthcare-to-improve-patient-outcomes/"
r = requests.get(URL)
soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
print(soup.prettify())
```

```python
def extract_article_text(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    article_text = ""

    # Extract article title
    title_element = soup.find('h1')
    if title_element:
        article_text += title_element.get_text().strip() + "\n\n"

    # Extract article text from the <div> tag with class="td-post-content" and tagdiv-type="text"
    article_element = soup.find('div', class_='td-post-content')
    if article_element:
        # Exclude text inside <div> tag with class="wp-block-preformatted"
        preformatted_elements = article_element.find_all('pre', class_='wp-block-preformatted')
        for preformatted_element in preformatted_elements:
            preformatted_element.extract()

        article_text += article_element.get_text().strip()

    return article_text.strip()
```

```python
for _,row in data.iterrows():
        url_id = row['URL_ID']
        url = row['URL']
        article_text = extract_article_text(url)

        # Save the extracted article as a text file
        file_name = f"{url_id}.txt"
        with open(file_name, 'w') as f:
            f.write(article_text)

        print(f"Article extracted and saved: {file_name}")
```

```python
## Import all the required libraries

# for tokenization
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
from nltk.tokenize import sent_tokenize

# for stopwords
from nltk.corpus import stopwords
import string
nltk.download('stopwords')

# for personal pronouns
import re
```

```python
### Load input data from the excel file
data
```

```python
# creating list to store the variables

url_ids_list = []
urls_list = []
positive_scores_list = []
negative_scores_list = []
polarity_scores_list = []
subjectivity_scores_list = []
avg_sentence_lengths_list = []
percentage_complex_words_list = []
fog_indices_list = []
avg_words_per_sentence_list = []
complex_word_counts_list = []
word_counts_list = []
syllables_per_word_list = []
personal_pronouns_list = []
avg_word_lengths_list = []
```

```python
### Using Stopwords

stop_words_files = ["StopWords_Auditor.txt", "StopWords_Currencies.txt", "StopWords_DatesandNumbers.txt","StopWords_Generic.txt","StopWords_GenericLong.txt","StopWords_Geographic.txt","StopWords_Names.txt"]  # List of file paths

stop_words = set()  # Initialize an empty set for stop words

for file_path in stop_words_files:
    with open(file_path, 'r', encoding='latin-1') as file:  # Specify the correct encoding
        words = file.read().split()  # Read the file and split its contents into words
        stop_words.update(words)  # Add the words to the stop words set

print(stop_words)
```

```python
### performing textual analysis
for index, row in data.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    text = extract_article_text(url)  # Use your existing function to extract article text

    ### Tokenization
    tokens = word_tokenize(text)
    cleaned_tokens = [token for token in tokens if token.lower() not in stop_words]

    ### Use of Master Dictionary
    positive_words = set()
    negative_words = set()

    positive_dictionary_path = '/content/positive-words.txt'
    negative_dictionary_path = '/content/negative-words.txt'

    with open(positive_dictionary_path, 'r') as positive_file:
        positive_dictionary = positive_file.read().splitlines()

    with open(negative_dictionary_path, 'r' , encoding='latin-1') as negative_file:
        negative_dictionary = negative_file.read().splitlines()

    for word in cleaned_tokens:
        if word.lower() in positive_dictionary and word.lower() not in stop_words:
            positive_words.add(word.lower())
        if word.lower() in negative_dictionary and word.lower() not in stop_words:
            negative_words.add(word.lower())

    ### Extracting Data Variables
    positive_score = len(positive_words)
    negative_score = len(negative_words)

    overall_sentiment = positive_score - negative_score

    polarity_score = (positive_score - negative_score)/ ((positive_score + negative_score) + 0.000001)

    subjectivity_score = (positive_score + negative_score)/((len(cleaned_tokens) + 0.000001))

    ### Analysis of Readability
    words = len(cleaned_tokens)

    sentences = len(sent_tokenize(text))

    def syllables_count(word):
        word = word.lower()
        count = 0
        vowels = "aeiouy"
        exceptions = ["es", "ed"]

        if word.endswith(tuple(exceptions)):
            return count

        if word[0] in vowels:
            count += 1

        for index in range(1, len(word)):
            if word[index] in vowels and word[index - 1] not in vowels:
                count += 1

        if word.endswith("e"):
            count -= 1

        if count == 0:
            count += 1

        return count

    syllable_count = sum(syllables_count(word) for word in cleaned_tokens)

    # Complex words
    complex_words = 0
    for word in cleaned_tokens:
        # Count the number of syllables in the word
        syllables = syllables_count(word)
        # Check if the word has more than two syllables
        if syllables > 2:
            complex_words += 1

    average_sentence_length = words/(sentences+0.0000001)
    percentage_complex_words = (complex_words / (words+0.0000001)) * 100
    fog_index = 0.4 * (average_sentence_length + percentage_complex_words)

    total_words = len(cleaned_tokens)
    total_sentences = len(sent_tokenize(text))

    average_words_per_sentence = total_words / (total_sentences+0.0000001)

    ### Word count using stopwords from NLTK library
    # Get the list of English stop words
    stop_words = set(stopwords.words('english'))

    # Remove punctuation marks from the set of string.punctuation
    translator = str.maketrans('', '', string.punctuation)

    # Remove stop words and punctuation marks from cleaned_tokens
    cleaned_tokens = [token.lower().translate(translator) for token in cleaned_tokens if token.lower() not in stop_words]

    # Count the number of words
    word_count = len(cleaned_tokens)

    ## Syllable count per word
    # ==> Using the already defined function

    # Personal Pronouns
    personal_pronouns = ["I", "we", "my", "ours", "us"]

    # Join the personal pronouns into a regex pattern
    pattern = r'\b(?:{})\b'.format('|'.join(personal_pronouns))

    # Search for the pattern in the text
    matches = re.findall(pattern, text, flags=re.IGNORECASE)

    # Count the number of personal pronouns
    personal_pronoun_count = len(matches)

    # Average Word Length
    total_characters = sum(len(word) for word in cleaned_tokens)
    total_words = len(cleaned_tokens)

    average_word_length = total_characters / (total_words+0.0000001)

    # Append computed values to lists
    url_ids_list.append(url_id)
    urls_list.append(url)
    positive_scores_list.append(positive_score)
    negative_scores_list.append(negative_score)
    polarity_scores_list.append(polarity_score)
    subjectivity_scores_list.append(subjectivity_score)
    avg_sentence_lengths_list.append(average_sentence_length)
    percentage_complex_words_list.append(percentage_complex_words)
    fog_indices_list.append(fog_index)
    avg_words_per_sentence_list.append(average_words_per_sentence)
    complex_word_counts_list.append(complex_words)
    word_counts_list.append(word_count)
    syllables_per_word_list.append(syllable_count)
    personal_pronouns_list.append(personal_pronoun_count)
    avg_word_lengths_list.append(average_word_length)
```

```python
# dataframe to save the output values

df_output = pd.DataFrame({
    'URL_ID': url_ids_list,
    'URL': urls_list,
    'POSITIVE SCORE': positive_scores_list,
    'NEGATIVE SCORE': negative_scores_list,
    'POLARITY SCORE': polarity_scores_list,
    'SUBJECTIVITY SCORE': subjectivity_scores_list,
    'AVG SENTENCE LENGTH': avg_sentence_lengths_list,
    'PERCENTAGE OF COMPLEX WORDS': percentage_complex_words_list,
    'FOG INDEX': fog_indices_list,
    'AVG NUMBER OF WORDS PER SENTENCE': avg_words_per_sentence_list,
    'COMPLEX WORD COUNT': complex_word_counts_list,
    'WORD COUNT': word_counts_list,
    'SYLLABLE PER WORD': syllables_per_word_list,
    'PERSONAL PRONOUNS': personal_pronouns_list,
    'AVG WORD LENGTH': avg_word_lengths_list
})
```
